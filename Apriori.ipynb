```python
import numpy as np # linear algebra
import pandas as pd

# !pip install apyori
# from apyori import apriori

from mlxtend.preprocessing import TransactionEncoder
```

```python
data = pd.read_csv('/content/gdrive/My Drive/DATASET_THESIS_UNDERGRAD/PUBLIC_PRIVATE_DATA_SPLIT_PUBLIC.csv')

df = data
df.head()
df
```

```python
df.drop(['Likes_University_Subject', 'No_Tuition_Fee_Struggle', 'Not_Bullied', 'No_Sexual_Harrassment', 'Good_Familial_Relationship', 'University_Subject_Less_Criticised', 'Low_Academic_Stress', 'Very_Cooperative_Faculties', 'Very_Cooperative_Classmates', 'PHQ_Minimal_Depression'], axis=1, inplace=True)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori

fritems = apriori(df, min_support=0.05, use_colnames=True)
fritems
```

```python
from mlxtend.frequent_patterns import association_rules
res = association_rules(fritems, metric='confidence', min_threshold=0.2)
res
```
