```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('/content/gdrive/My Drive/DATASET_THESIS_UNDERGRAD/PUBLIC_PRIVATE_DATA_FINAL.csv')
data
```

```python
datanew = data.fillna(0)
```

```python
#Male-Female 1-hot encoding split

dummies = pd.get_dummies(datanew.Sex)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Sex'], axis='columns')

datanew.rename(columns = {2: 'Female', 1: 'Male'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Residence)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Residence'], axis='columns')

datanew.rename(columns = {1: 'Family_Residence', 2: 'Hostel_Residence', 3:'Relative_Residence'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Romantic_Relationship)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Romantic_Relationship'], axis='columns')

datanew.rename(columns = {1: 'In_a_Relationship', 2: 'Single', 3:'Complicated_Relationship'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Number_Of_Friends)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Number_Of_Friends'], axis='columns')

datanew.rename(columns = {1: 'No_Friend', 2: 'Minimal_Friends', 3:'Group_Of_Friends', 4:'Multiple_Group_Of_Friends'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Frequency_Of_Going_Out)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Frequency_Of_Going_Out'], axis='columns')

datanew.rename(columns = {1: 'Rare_Social_Interaction', 2: 'Weekly_Social_Interaction', 3:'Daily_Social_Interaction'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Choice_Of_Subject)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Choice_Of_Subject'], axis='columns')

datanew.rename(columns = {1: 'Self_Chosen_Subject', 2: 'Subject_by_Merit_Position', 3:'Subject_by_Family_Pressure'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Subject_Likeliness)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Subject_Likeliness'], axis='columns')

datanew.rename(columns = {1: 'Likes_University_Subject', 2: 'Dislikes_University_Subject'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.University_Facility, drop_first=True)

merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['University_Facility'], axis='columns')

datanew.rename(columns = {2: 'Dissatisfied_With_University_Facilities'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Session_Jam)
dummies.drop(2, inplace=True, axis=1)

merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Session_Jam'], axis='columns')

datanew.rename(columns = {1: 'Faced_Session_Jam'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Tuition_Fee_Struggle)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Tuition_Fee_Struggle'], axis='columns')

datanew.rename(columns = {1: 'No_Tuition_Fee_Struggle', 2: 'Extreme_Tuition_Fee_Struggle', 3:'Moderate_Tuition_Fee_Struggle'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Bullied)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Bullied'], axis='columns')

datanew.rename(columns = {1: 'Bullied_as_as_Freshman', 2: 'Bullied_Often', 3:'Not_Bullied'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Sexual_Harrassment)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Sexual_Harrassment'], axis='columns')

datanew.rename(columns = {1: 'No_Sexual_Harrassment', 2: 'Indirect_Sexual_Harrassment', 3:'Direct_Sexual_Harrassment'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Age)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Age'], axis='columns')

datanew.rename(columns = {1: 'Age_20-22', 2: 'Age_22-24', 3:'Age_24+'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Gadget_Use_Hour)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Gadget_Use_Hour'], axis='columns')

datanew.rename(columns = {1: 'Gadget_Usage_0-5', 2: 'Gadget_Usage_5-10', 3:'Gadget_Usage_10-15', 4:'Gadget_Usage_15-20', 5:'Gadget_Usage_20+'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Study_Hour)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Study_Hour'], axis='columns')

datanew.rename(columns = {1: 'Study_Hour_0-5', 2: 'Study_Hour_5-10', 3:'Study_Hour_10-15', 4:'Study_Hour_15-20', 5:'Study_Hour_20+'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Smoke_Habit)
merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Smoke_Habit'], axis='columns')

datanew.rename(columns = {1: 'Regular_Smoker', 2: 'Irregular_Smoker', 3:'Non_Smoker'}, inplace = True)

datanew
```

```python
dummies = pd.get_dummies(datanew.Drugs_Intake)
dummies.drop(2, inplace=True, axis=1)
# dummies

merged = pd.concat([datanew, dummies], axis='columns')
datanew = merged.drop(['Drugs_Intake'], axis='columns')

datanew.rename(columns = {1: 'Drug_Intake'}, inplace = True)

datanew
```
