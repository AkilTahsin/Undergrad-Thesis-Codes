```python
# import necessary modules 
import pandas  as pd
import matplotlib.pyplot as plt
import numpy as np

# load the data set
datanew = pd.read_csv('/content/gdrive/My Drive/DATASET_THESIS_UNDERGRAD/PUBLIC_PRIVATE_DATA_PRE_PROCESSED.csv')
# print(data)

# print info about columns in the dataframe
print(datanew.info())
```

```python
#divides data into X and y dataframes

X = datanew.drop(['University_In_Dhaka'], axis='columns')
y = datanew['University_In_Dhaka']

# y.value_counts()
X
```

```python
# using 1-hot encoding to convert y into numerical value dataset
# University Out of Dhaka -> 1, University Inside Dhaka -> 0
y = pd.get_dummies(y, drop_first=True)

y.sample(10)
# y.value_counts()
# y.columns
```

```python
# import SMOTE module from imblearn library
# pip install imblearn (if you don't have imblearn in your system)

from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state = 42)
X_sm, y_sm = sm.fit_sample(X, y.values.ravel())
```

```python
X_sm_df = pd.DataFrame(X_sm, columns=X.columns)
y_sm_df = pd.DataFrame(y_sm, columns =y.columns)

y_sm_df
```

```python
datasm = np.concatenate((X_sm_df, y_sm_df), axis = 1)
datasm_df = pd.DataFrame(datasm, columns =['PHQ_1', 'PHQ_2', 'PHQ_3', 'PHQ_4', 'PHQ_5', 'PHQ_6', 'PHQ_7', 'PHQ_8',
       'PHQ_9', 'Age', 'Sex', 'Area', 'Residence', 'Gadget_Use_Hour',
       'Social_Media_Hour', 'Study_Hour', 'Smoke_Habit', 'Drugs_Intake',
       'Relationship_With_Family', 'Romantic_Relationship',
       'Number_Of_Friends', 'Frequency_Of_Going_Out', 'Public_or_Private',
       'Choice_Of_Subject', 'Subject_Likeliness',
       'University_Facility', 'Session_Jam', 'Tuition_Fee_Struggle',
       'University_Subject_Criticism', 'Academic_Stress',
       'Faculty_Cooperation', 'Classmate_Cooperation',
       'Extra_Curricular_Activity', 'Bullied', 'Sexual_Harrassment', 'University_out_of_Dhaka'])

datasm_df
```

```python
datasm_df.to_csv('PUBLIC_PRIVATE_DATA_SMOTE_University_Location.csv', index=False)



!cp PUBLIC_PRIVATE_DATA_SMOTE_University_Location.csv "/content/gdrive/My Drive/DATASET_THESIS_UNDERGRAD/"
```
